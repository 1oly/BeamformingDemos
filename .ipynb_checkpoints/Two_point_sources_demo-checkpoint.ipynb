# Beamforming demo
Interactive plotting of delay-and-sum beamforming.  
These measurements were conducted at the large anechoic chamber at [DTU Elektro](https://www.elektro.dtu.dk) with a microphone array and two point-like sources.  

The measurements were post-processed and beamforming outputs were stored for 40 frequency-bins with a binwidth of 256 Hz.

### How to use this interactive demo
This is a [Jupyter notebook](http://nbviewer.jupyter.org/github/jupyter/notebook/blob/master/docs/source/examples/Notebook/Notebook%20Basics.ipynb), goto a cell below and press  
```Ctrl+Enter``` (Windows) or  
```Shift+Enter``` (Mac)  

Below is basic code for running demo using the [Julia language](http://julialang.org).

````julia
using JLD2, Interact, Plots
@load "two_point_sources.jld2" b_two
f = 0:256:9984;
````

````julia
@manipulate for i = 1:40
    contourf(-1.0:0.025:1.0,-1.0:0.025:1.0,10*log10.(b_two[:,:,i]'/2e-5^2),
        aspectratio=1,
        xlabel="x [m]",
        ylabel="y [m]",
        title="f = $(f[i]) Hz",
        clim=(30,50))
end
````

Interact with the plot by pulling the slider to change frequency of the beamformer
